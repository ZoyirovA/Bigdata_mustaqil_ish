## **1-qadam. Kutubxonalarni chaqirish**

Energetika boâ€˜yicha maâ€™lumotlarni tahlil qilish uchun bizga pandas va matplotlib kerak boâ€˜ladi.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Grafiklar Colab ichida koâ€˜rinishi uchun
%matplotlib inline
```

## **2-qadam. Faylni yuklash va koâ€˜rib chiqish**

Excel faylingizni Colabâ€™ga yuklab olamiz va dastlabki qatorlarini tekshiramiz.

```python
df = pd.read_excel("sample_data/energy.xlsx")  # fayl nomini mos ravishda yozing

# Birinchi 5 qatorni chiqaramiz
df.head()
```

**3-qadam. Maâ€™lumotlarni tartiblash**

Yillar ketma-ketligida grafik chizish uchun `Year` boâ€˜yicha tartiblaymiz.


```python
df = df.sort_values("Year")
```

**4-qadam. Qayta tiklanuvchi elektr energiyasi ulushi**

Bu grafikda qayta tiklanuvchi manbalardan olingan elektr energiyasi ulushi (%) tasvirlanadi.


```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Renewable Electricity Output (% of total)"],
         marker="o", color="green")
plt.title("Qayta tiklanuvchi elektr energiyasi ulushi (%)")
plt.xlabel("Yil")
plt.ylabel("Ulush, %")
plt.grid(True)
plt.show()
```

**5-qadam. Energiya isteâ€™moli (aholi boshiga)**

Bu yerda bir kishiga toâ€˜gâ€˜ri keladigan energiya isteâ€™moli (kg neft ekvivalenti) koâ€˜rsatiladi.


```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"],
         df["Energy Use (kg of oil equivalent per capita)"],
         marker="s", color="blue")
plt.title("Energiya isteâ€™moli (aholi boshiga, kg neft ekvivalenti)")
plt.xlabel("Yil")
plt.ylabel("kg neft ekv. / kishi")
plt.grid(True)
plt.show()
```

**6-qadam. Koâ€˜mir va fosil yoqilgâ€˜ilar ulushi**

Energetika balansida koâ€˜mir va umumiy fosil yoqilgâ€˜ilarning ulushi (%) taqqoslanadi.


```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"],
         df["Energy Use from Coal (% of total)"],
         marker="^", label="Koâ€˜mir (%)")
plt.plot(df["Year"],
         df["Energy Use from Fossil Fuels (% of total)"],
         marker="o", label="Fosil yoqilgâ€˜i (%)")
plt.title("Energiya manbalari ulushi")
plt.xlabel("Yil")
plt.ylabel("Ulush, %")
plt.legend()
plt.grid(True)
plt.show()
```

**7-qadam. Energiya intensivligi**

Bu koâ€˜rsatkich iqtisodiy oâ€˜sish uchun qancha energiya sarflanishini koâ€˜rsatadi (MJ / YaIM birligi).


```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"],
         df["Energy Intensity (MJ per GDP unit)"],
         marker="d", color="red")
plt.title("Energiya intensivligi (MJ / YaIM birlik)")
plt.xlabel("Yil")
plt.ylabel("MJ / GDP")
plt.grid(True)
plt.show()
```

**8-qadam. Toza yoqilgâ€˜i va texnologiyalarga kirish**

Aholining toza yoqilgâ€˜i va texnologiyalarga ulanish darajasi (%) tasvirlanadi.


```python
plt.figure(figsize=(10,5))
plt.plot(df["Year"],
         df["Access to Clean Fuels (% of population)"],
         marker="o", color="orange")
plt.title("Toza yoqilgâ€˜i va texnologiyalarga kirish (%)")
plt.xlabel("Yil")
plt.ylabel("Ulush, %")
plt.grid(True)
plt.show()
```
